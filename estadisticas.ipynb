```python
import pandas as pd

def analizar_trafico(ruta):
    df = pd.read_parquet(ruta, engine='fastparquet')

    total_paquetes = len(df)
    total_bytes = int(df['Length'].sum())
    tamaño_medio = total_bytes / total_paquetes if total_paquetes > 0 else 0
    protocolo_frecuente = df['Protocolo'].mode()[0] if not df['Protocolo'].empty else "N/A"
    protocolos_distintos = df['Protocolo'].nunique()

    tiempo_total = df['Trelativo'].max()
    tasa_llegada = total_paquetes / tiempo_total if tiempo_total > 0 else 0

    resultados = {
        "Métrica": [
            "Total paquetes capturados",
            "Tasa media de llegada (paq/s)",
            "Tamaño medio paquete (Bytes)",
            "Protocolo más frecuente",
            "Total de bytes",
            "Número de protocolos distintos"
        ],
        "Valor": [
            total_paquetes,
            f"{tasa_llegada:.2f}",
            f"{tamaño_medio:.2f}",
            protocolo_frecuente,
            total_bytes,
            protocolos_distintos
        ]
    }

    tabla_resultados = pd.DataFrame(resultados)
    return tabla_resultados

tabla_real = analizar_trafico("traficoligero.parquet")
print("\n--- RESULTADOS DEL ARCHIVO LIGERO ---")
print(tabla_real.to_string(index=False))

print("-" * 50)

tabla_real2 = analizar_trafico("traficopesado.parquet")
print("\n--- RESULTADOS DEL ARCHIVO PESADO ---")
print(tabla_real2.to_string(index=False))
```
